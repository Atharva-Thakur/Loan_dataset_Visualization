```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/loan_applications.csv')
```

```python
df['index_column'] = df.index
df.head()
```

```python
# Histogram
plt.rcParams["figure.figsize"] = [14, 5]
plt.title("Distribution of Credit Scores")
plt.hist(df['credit_score'], bins=40)
plt.show()
```

```python

for r in df.race.unique():
  print(r)
  print((df.race==r).value_counts())
```

```python
y = np.array([df['race'].value_counts()])
pieLabels = ['White','Asian','Hispanic','Black']
print(pieLabels)
print(y)
```

```python
plt.pie(y, labels = pieLabels)
plt.show() 
```

```python
# bar graph showing a comparison of between races and will_default
crosstab_result = pd.crosstab(index=df['will_default'], columns=df['race'])
crosstab_result.plot.bar()
```

```python
# Ploting credit scores for each group using line graph
for race in sorted(df['race'].unique()):
  plt.figure(figsize=(30,4))
  plt.title(race)
  plt.plot(df[df['race']==race]["index_column"],df[df["race"] == race]['credit_score'])
  plt.show()
```

```python
# as we can see that it is not efficient to plot credit scores for each group using line graph so we would try using histogram for this
for race in sorted(df['race'].unique()):
  plt.figure(figsize=(30,4))
  plt.title(race)
  plt.hist(df[df["race"] == race]['credit_score'], bins=40)
  plt.show()
```
